```
# 3. Inserisci la tua API Key Regolo.ai
import os
os.makedirs('.streamlit', exist_ok=True)

api_key = input('Incolla la tua REGOLO_API_KEY: ')

with open('.streamlit/secrets.toml', 'w') as f:
    f.write(f'REGOLO_API_KEY = "{api_key}"')

print('Secrets salvati.')
```

```
# 4. Configura ngrok (token da https://dashboard.ngrok.com/get-started/your-authtoken)
from pyngrok import ngrok

ngrok_token = input('Incolla il tuo ngrok authtoken: ')
ngrok.set_auth_token(ngrok_token)
print('Token configurato.')
```

```
# 5. Avvia Streamlit e apri il tunnel
import subprocess, threading, time

def run_streamlit():
    subprocess.run(['streamlit', 'run', 'app.py',
                    '--server.port', '8501',
                    '--server.headless', 'true'])

t = threading.Thread(target=run_streamlit, daemon=True)
t.start()
time.sleep(4)

public_url = ngrok.connect(8501)
print(f'\n*** Apri l\'app qui: {public_url} ***')
```
